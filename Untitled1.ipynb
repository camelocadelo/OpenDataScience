```python
def age_category(age):
    if age < 30:
        return 1
    elif age < 55:
        return 2
    else:
        return 3
```

```python
data['Age category'] = data['Age'].apply(age_category)
```

```python
def new_age_category(age):
    if age<30:
        return 1
    elif (age>= 30) & (age<=60):
        return 2
    elif age>60:
        return 3
```

```python
data['New age category'] = data['Age'].apply(new_age_category)
```
